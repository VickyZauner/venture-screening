```python
# tune dictionairies
def categorize(column):
    for company in column:
        if word in company in ['genomic', 'gene', 'genes', 'instability', 'instable']:
            return'genomic instabilty'
        if word in company in ['senescence', 'senescent', 'Senescence', 'Senescent']:
            return 'cellular senescence'
        if word in company in ['telomere', 'telomerase', 'telomere attrition', 'attition']:
            return 'telomere attriton'
        if word in company in ['epigenetic alterations', 'epigene', 'epigenetic']:
            return 'epigenetic alterations'
        if word in company in ['loss of proteostasis', 'proteostasis', 'protein']:
            return 'loss of proteostasis'
        if word in company in ['deregulated nutrient sensing', 'nutrient sensing']:
            return 'deregulated nutrient sensing'
        if word in company in ['mitochondrial dysfunction', 'mitochondrial', 'mitochondria']:
            return 'mitochondrial dysfunction'
        if word in company in ['stem cell exhaustion', 'stem cell']:
            return 'stem cell exhaustion'
        else:
            return 'undefined'
        
```
